```python
import requests
```

```python
r = requests.get('https://wall.alphacoders.com/by_category.php?id=32&name=Video+Game+Wallpapers&filter=4K+Ultra+HD')
print(r)
```

```python
#to know all function
print(dir(r))
```

```python
#to know all functions in detail
print(help(r))
```

```python
#text which is available in help(r)
print(r.text)
```

```python
#this is also available in help(r)
#check it for more information
print(r.content)
```

```python
with open('request','wb') as f:
    f.write(r.content)
```

```python
#it gives http code search for http code to know more
print(r.status_code)
#as per our output 200 means successful
```

```python
#ok checks weather the status is not 400 sinsce 
#400 in http is error status
print(r.ok)
```

```python
#it gives the site information
print(r.headers)
```

```python
payload = {'page':2,'count':25}
r = requests.get('https://httpbin.org/get',params = payload)
print(r.text)
```

```python
print(r.url)
```
